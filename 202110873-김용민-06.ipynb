```python
import numpy as np                                                                #    numpy import
import networkx as nx                                                             #    networkx import
import matplotlib.pyplot as plt                                                   #    matplotlib.pyplot import

def generate_adjacency_matrix(n):                                                 # n x n    
    matrix = np.random.randint(2, size=(n, n))                                    #    randint  , matrix  .
    return matrix                                                                 #   .

def analyze_graph(matrix):                                                        #    ,  ,  ,      
    G = nx.from_numpy_array(matrix, create_using=nx.DiGraph())                    # from_numpy_array        G .
    loops = sum(matrix[i][i] for i in range(len(matrix)))                         #    loops .
    degrees = dict(G.degree())                                                    #     degrees .
    edges = G.number_of_edges()                                                   #    edges .

    return G, loops, degrees, edges                                               # G, loops, degrees, edges .

def print_graph_info(matrix, loops, degrees, edges):                              # G, loops, degrees, edges   
    print(" :")                                                           # " : " .
    print(matrix)                                                                 #  G .
    print("\n :", edges)                                                  #   edges .
    print(" :", loops)                                                    #   loops .
    print("  :")                                                      # "   : " .
    for node, degree in degrees.items():                                          #      for  .
        print(f"Node {node}: {degree}")                                           #    degree .

def plot_graph(G):                                                                # G    
    pos = nx.spring_layout(G)                                                     # G   pos .
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, arrowsize=20)         #     draw   G .
    plt.show()                                                                    #    .

for n in [3, 4, 5]:
    print(f"\n--- n={n}  ,   ---")                                  # n = 3, 4, 5     
    matrix = generate_adjacency_matrix(n)                                         # n    
    G, loops, degrees, edges = analyze_graph(matrix)                              # ,  ,  ,    .
    print_graph_info(matrix, loops, degrees, edges)                               #   .
    plot_graph(G)                                                                 #   .


#         .
# np.random.randint(2, size=(n, n))   0,1      ,
#      ,  ,        .

```
