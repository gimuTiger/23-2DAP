```python
import numpy as np                                                                # 수리 연산을 위한 numpy import
import networkx as nx                                                             # 그래프를 만들기 위한 networkx import
import matplotlib.pyplot as plt                                                   # 그래프를 출력하기 위한 matplotlib.pyplot import

def generate_adjacency_matrix(n):                                                 # n x n 인접행렬을 생성하는 함수 정의
    matrix = np.random.randint(2, size=(n, n))                                    # 임의로 생성하기 위해 randint 함수를 사용하고, matrix 에 저장한다.
    return matrix                                                                 # 생성된 행렬을 반환한다.

def analyze_graph(matrix):                                                        # 생성된 행렬의 그래프를 그리고, 간선의 수, 루프의 수, 각 정점의 차수를 저장하는 함수 정의
    G = nx.from_numpy_array(matrix, create_using=nx.DiGraph())                    # from_numpy_array 함수를 이용하여 매개변수로 받은 행렬의 그래프를 그리고 G에 저장한다.
    loops = sum(matrix[i][i] for i in range(len(matrix)))                         # 루프의 수를 계산하여 loops에 저장한다.
    degrees = dict(G.degree())                                                    # 각 정점의 차수를 계산하여 degrees에 저장한다.
    edges = G.number_of_edges()                                                   # 간선의 수를 계산하여 edges에 저장한다.

    return G, loops, degrees, edges                                               # G, loops, degrees, edges를 반환한다.

def print_graph_info(matrix, loops, degrees, edges):                              # G, loops, degrees, edges를 출력하는 함수를 정의
    print("인접 행렬:")                                                           # "인접행렬 : " 출력한다.
    print(matrix)                                                                 # 행렬 G를 출력한다.
    print("\n간선의 수:", edges)                                                  # 간선의 수 edges를 출력한다.
    print("루프의 수:", loops)                                                    # 루프의 수 loops를 출력한다.
    print("각 정점의 차수:")                                                      # "각 정점의 차수 : " 출력한다.
    for node, degree in degrees.items():                                          # 각 정점의 차수를 출력하기 위해 for 함수를 사용한다.
        print(f"Node {node}: {degree}")                                           # 각 정점의 차수 degree를 출력한다.

def plot_graph(G):                                                                # G를 그래프로 그리는 함수를 정의
    pos = nx.spring_layout(G)                                                     # G의 위치를 설정하여 pos에 저장한다.
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, arrowsize=20)         # 정점과 간선의 그래프를 그려주는 draw 함수를 사용하여 G를 그린다.
    plt.show()                                                                    # 그린 그래프를 화면에 출력한다.

for n in [3, 4, 5]:
    print(f"\n--- n={n} 일 때, 그래프 분석 ---")                                  # n = 3, 4, 5 인 경우에 대하여 각각 출력
    matrix = generate_adjacency_matrix(n)                                         # n에 대한 인접 행렬 생성
    G, loops, degrees, edges = analyze_graph(matrix)                              # 그래프, 간선의 수, 루프의 수, 각 정점의 차수를 구한다.
    print_graph_info(matrix, loops, degrees, edges)                               # 구한 것을 출력한다.
    plot_graph(G)                                                                 # 그래프를 화면에 출력한다.


# 실행 결과를 검토한 결과 의도한 결과를 얻은 것으로 생각됩니다.
# np.random.randint(2, size=(n, n)) 를 사용하여 0,1로 무작위로 채워진 행렬을 만들 수 있었으며,
# 그래프를 분석하는 함수를 사용하여 간선의 수, 루프의 수, 각 정점의 차수 등을 정확히 구할 수 있었습니다.

```
